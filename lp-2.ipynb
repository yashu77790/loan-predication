```python
import pandas as pd
import matplotlib.pyplot as plt

Loan_data = pd.read_csv('/content/loan_data_set.csv') # replace 'loan_data.csv' with the actual name of your loan data file

# assuming you have created an AxesSubplot object named 'plot'
plot = Loan_data.boxplot(column='LoanAmount', by='Loan_Status')
plot.set_title('Loan Amount vs Loan Status')
plot.set_xlabel('Loan Status')
plot.set_ylabel('Loan Amount')
plt.show() # add this line to display the plot
```

```python
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
matrix= Loan_data.corr() 
fig, ax = plt.subplots(figsize=(9,6))
cmap = sns.dark_palette("#fd6925", as_cmap=True)
fig.tight_layout()
HeatM = sns.heatmap(matrix,vmax=.8,square=True, cmap=cmap)
plt.rcParams['figure.figsize']=(10,10)
fig = HeatM.get_figure()
fig.savefig('output.png', transparent=True)
```
